# Concat与Append操作

````python
# 首先导入包
import numpy as np
import pandas as pd
````

````python
def make_df(cols, ind):
    data = {
        c : [str(c) + str(i) for i in ind]
        for c in cols
    }
    return pd.DataFrame(data, ind)

make_df('ABC', range(3))
````

# 实例：美国各州的统计数据

````python
# 下载数据
# !curl -O https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-population.csv
# !curl -O https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-areas.csv
# !curl -O https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-abbrevs.csv
````

````python
pop = pd.read_csv('state-population.csv')
areas = pd.read_csv('state-areas.csv')
abbrevs = pd.read_csv('state-abbrevs.csv')
print(pop.head())
print(areas.head())
print(abbrevs.head())
````

首先将pop的`state/region`列与abbrevs的`abbreviation`列进行合并， 还需要通过`how='outer'`确保数据没有丢失

````python
merged = pd.merge(pop, abbrevs, how='outer', left_on='state/region', right_on='abbreviation')
merged = merged.drop('abbreviation', 1) # 丢弃重复信息
merged.head()
````

````python
merged.isnull().any()
````

````python
merged[merged['population'].isnull()].head()
````

````python
# 看哪个州有缺失
merged.loc[merged['state'].isnull(), 'state/region'].unique()
````

````python
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'
merged.isnull().any()
````

将两个数据集共同的state列来合并

````python
final = pd.merge(merged, areas, on='state', how='left')
final.head()
````

检查一下哪些列有缺失值

````python
final.isnull().any()
````

看看是哪个地区的面积缺失

````python
final['state'][final['area (sq. mi)'].isnull()].unique()
````

````python
final.dropna(inplace=True)
final.head()
````

````python
data2010 = final.query("year == 2010 & ages == 'total'")
data2010.head()
````

计算人口密度按序排序。对索引重置，然后计算结果：

````python
data2010.set_index('state', inplace=True)
density = data2010['population'] / data2010['area (sq. mi)']

density.sort_values(ascending=False, inplace=True)
density.head()
````

# 高性能Pandas：eval()与query()

## 用Pandas.eval()实现高性能运算

````python
import numpy as np
import pandas as pd
nrows, ncols = 100000, 100
rng = np.random.RandomState(42)
df1, df2, df3, df4 =(pd.DataFrame(rng.rand(nrows, ncols)) for i in range(4))
````

````python
%timeit df1 + df2 + df3 + df4
````

````python
%timeit pd.eval('df1 + df2 + df3 + df4')
````

````python
np.allclose(df1 + df2 + df3 + df4, pd.eval('df1 + df2 + df3 + df4'))
````

pd.eval()支持的运算

````python
df1, df2, df3, df4, df5 = (pd.DataFrame(rng.randint(0, 1000, (100, 3))) for i in range(5))
````

（1）算术运算符

````python
result1 = -df1 * df2 / (df3 + df4) - df5
result2 = pd.eval('-df1 * df2 / (df3 + df4) -df5')
np.allclose(result1, result2)
````

（2）比较运算符

````python
result1 = (df1 < df2) & (df2 <= df3) & (df3 != df4)
result2 = pd.eval('(df1 < df2) & (df2 <= df3) & (df3 != df4)')
np.allclose(result1, result2)
````

（3）位运算符

````python
result1 = (df1 < 0.5) & (df2 < 0.5) | (df3 < df4)
result2 = pd.eval('(df1 < 0.5) & (df2 < 0.5) | (df3 < df4)')
np.allclose(result1, result2)
````

````python
result3 = pd.eval('(df1 < 0.5) and (df2 < 0.5) or (df3 < df4)')
np.allclose(result1, result3)
````

（4）对象属性与索引

````python
result1 = df2.T[0] + df3.iloc[1]
result2 = pd.eval('df2.T[0] + df3.iloc[1]')
np.allclose(result1, result2)
````
